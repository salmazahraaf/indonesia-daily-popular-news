Import necessary package

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## Detik

```python
req = requests.get('https://www.detik.com/terpopuler', params={'tag_from': 'wp_cb_mostPopular_more'})
print(req)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
#print(soup)
```

```python
popular_area = soup.find(attrs={'class': 'grid-row list-content'})
content_item = popular_area.find_all(attrs={'class': 'list-content__item'})
```

```python
detik_list = [[f"{index.find_all(attrs={'class': 'media__title'})[0].find('a').text}", 
               f"{index.find_all(attrs={'class': 'media__date'})[0].find('span')['title']}",
               f"{index.find_all(attrs={'class': 'media__title'})[0].find('a')['href']}"] for index in content_item]
```

```python
detik_df = pd.DataFrame(detik_list,columns=["title", "time", "url"])
detik_df['media'] = 'detik'
detik_df
```

## Liputan 6

```python
req = requests.get('https://www.liputan6.com/indeks/terpopuler/hari-ini')#, params={'tag_from': 'wp_cb_mostPopular_more'})
print(req)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
#print(soup)
```

```python
popular_area = soup.find(attrs={'class': 'articles--list articles--list_rows'})
content_item = popular_area.find_all(attrs={'class': 'articles--rows--item'})
```

```python
liputan6_list = [[index.find_all(attrs={'class': 'articles--rows--item__title'})[0].find('a')['title'], 
               index.find_all(attrs={'class': 'articles--rows--item__time timeago'})[0]['datetime'],
               index.find_all(attrs={'class': 'articles--rows--item__title'})[0].find('a')['href']] for index in content_item]
```

```python
liputan6_df = pd.DataFrame(liputan6_list,columns=["title", "time", "url"])
liputan6_df['time'] = liputan6_df['time'].str.replace('T', ' ')
liputan6_df['media'] = 'liputan 6'
liputan6_df
```

## Kompas

```python
req = requests.get('https://www.kompas.com')
print(req)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
#print(soup)
```

```python
popular_area = soup.find(attrs={'class': 'most__wrap clearfix'})
content_item = popular_area.find_all(attrs={'class': 'most__list clearfix'})
```

```python
kompas_list = [[index.find_all(attrs={'class': 'most__title'})[0].text, 
               index.find_all(attrs={'class': 'most__link'})[0]['href']] for index in content_item]
```

```python
kompas_df = pd.DataFrame(kompas_list,columns=["title", "url"])
kompas_df['media'] = 'kompas'
kompas_df
```

## Tribun

```python
req = requests.get('https://www.tribunnews.com/populer?section=&type=1d', params={'tag_from': 'wp_cb_mostPopular_more'})
print(req)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
#print(soup)
```

```python
popular_area = soup.find(attrs={'class': 'lsi pt10 pb10'})
content_item = popular_area.find_all(attrs={'class': 'pt5 pb20'})
```

```python
tribun_list = [[index.find_all(attrs={'class': 'fbo f16 pt5 ln19'})[0].find('a')['title'], 
               index.find_all(attrs={'class': 'grey pt5'})[0].text,
               index.find_all(attrs={'class': 'fbo f16 pt5 ln19'})[0].find('a')['href']] for index in content_item]
```

```python
tribun_df = pd.DataFrame(tribun_list,columns=["title","time", "url"])
tribun_df['media'] = 'tribun' 
tribun_df
```

Save your result to xlsx file

```python
with pd.ExcelWriter('trend-news.xlsx') as writer:  
    detik_df.to_excel(writer, sheet_name='detik')
    liputan6_df.to_excel(writer, sheet_name='liputan6')
    tribun_df.to_excel(writer, sheet_name='tribun')
    kompas_df.to_excel(writer, sheet_name='kompas')
```
